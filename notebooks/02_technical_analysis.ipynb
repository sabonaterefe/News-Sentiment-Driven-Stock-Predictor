```python
from data_loader import load_all_data
from technical_analysis import calculate_technical_indicators
import matplotlib.pyplot as plt

# Load and process data
_, stocks = load_all_data()
stocks = calculate_technical_indicators(stocks)

# Example visualization for AAPL
aapl = stocks['AAPL']
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Price and MAs
aapl[['Close', 'MA_20', 'MA_50']].plot(ax=ax1)
ax1.set_title('AAPL Price and Moving Averages')

# RSI
aapl['RSI_14'].plot(ax=ax2, color='orange')
ax2.axhline(70, color='r', linestyle='--')
ax2.axhline(30, color='g', linestyle='--')
ax2.set_title('Relative Strength Index')

plt.tight_layout()
plt.show()
```
