```python
from data_loader import load_all_data
import matplotlib.pyplot as plt

# Load data
ratings, stocks = load_all_data()

# Basic stats
print(f"Total ratings: {len(ratings)}")
print(f"Date range: {ratings['date'].min()} to {ratings['date'].max()}")
print(f"Stocks available: {list(stocks.keys())}")

# Plot article count over time
ratings.set_index('date').resample('W').size().plot(title='Articles per Week')
plt.show()
```
